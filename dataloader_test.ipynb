```python
import json

def find_expressions(img_idx):
    with open('./output/expressions.json', 'r') as f:
        data = json.load(f)

    for item in data:
        if img_idx in item['img_idxs']:
            return item['img_idxs'], item['expressions']

    return None, None

# 测试函数
img_idx = 5
img_idxs, expressions = find_expressions(img_idx)
print('img_idxs:', img_idxs)
print('expressions:', expressions)
```
